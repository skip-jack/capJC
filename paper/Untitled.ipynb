```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('E:/capJC/paper/original.csv')

# Check the first few rows of the dataframe
print(df.head())

# Function to create a pie chart
def create_pie_chart(data, title, labels=None):
    plt.figure(figsize=(8, 8))
    plt.pie(data.value_counts(), labels=labels if labels else data.value_counts().index, autopct='%1.1f%%')
    plt.title(title)
    plt.show()

# Create pie charts
create_pie_chart(df['Gender'], 'Gender Distribution', labels=['Male', 'Female'])
create_pie_chart(df['Race'], 'Race Distribution', labels=['White', 'Black', 'Asian', 'Indian', 'Others'])

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('original.csv')

# Define age groups
bins = [0, 2, 4, 13, 19, 25, 35, 45, 55, 65, 75, 85, 95, 105, 115]
labels = ['0-2', '3-4', '5-13', '14-19', '20-25', '26-35', '36-45', '46-55', '56-65', '66-75', '76-85', '86-95', '96-105', '106+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Function to create a pie chart
def create_pie_chart(data, title, labels=None):
    plt.figure(figsize=(10, 10))
    plt.pie(data.value_counts(sort=False), labels=labels if labels else data.value_counts(sort=False).index, autopct='%1.1f%%')
    plt.title(title)
    plt.show()

# Create pie chart for age groups
create_pie_chart(df['AgeGroup'], 'Age Group Distribution')
```
