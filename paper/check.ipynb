```python
# Re-running the image processing code with NumPy imported.

sample_img_paths = img_paths[:100]

# Re-initialize lists to hold image data and labels
X = []  # List to hold image data
Y = []  # List to hold labels (age and gender)
```

```python
# Re-processing the images
import cv2
import numpy as np

for path in sample_img_paths:
    img = cv2.imread(path)  # Read the image
    X.append(img)  # Append the processed image to the list X
    
    # Extract age and gender information from the filename
    y = os.path.basename(path).split("_")[0:2]
    Y.append(y)

# Convert lists to numpy arrays
X = np.array(X)
Y = np.array(Y)

# Returning the shape of arrays to verify the processing
X.shape, Y.shape
```

```python
# Splitting the remaining paths into a test set
# Assuming the first 100 were for training, we'll take the next 20 for testing to maintain an 80-20 split
test_img_paths = img_paths[100:120]

# Initialize lists to hold test image data and labels
X_test = []  # List to hold test image data
Y_test = []  # List to hold test labels (age and gender)

# Processing the test images
for path in test_img_paths:
    img = cv2.imread(path)  # Read the image
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    img = cv2.resize(img, (256, 256))  # Resize to 256x256
    img = np.array(img) / 255.0  # Normalize the pixel values
    X_test.append(img)  # Append the processed image to the list X_test
    
    # Extract age and gender information from the filename
    y = os.path.basename(path).split("_")[0:2]
    Y_test.append(y)
```
